```python
import warnings
import numpy as np
from scipy import signal
from ipywidgets import Button, IntSlider, Output, Layout, VBox, HBox, Dropdown, AppLayout, Label
import matplotlib.pyplot as plt
from IPython.display import Audio, display, clear_output

from collections import namedtuple, OrderedDict

%matplotlib inline

continuous_update = False

import asyncio

class Timer:
    def __init__(self, timeout, callback):
        self._timeout = timeout
        self._callback = callback
        self._task = asyncio.ensure_future(self._job())

    async def _job(self):
        await asyncio.sleep(self._timeout)
        self._callback()

    def cancel(self):
        self._task.cancel()

def debounce(wait):
    """ Decorator that will postpone a function's
        execution until after `wait` seconds
        have elapsed since the last time it was invoked. """
    def decorator(fn):
        if continuous_update:
            timer = None
            def debounced(*args, **kwargs):
                nonlocal timer
                def call_it():
                    fn(*args, **kwargs)
                if timer is not None:
                    timer.cancel()
                timer = Timer(wait, call_it)
            return debounced
        else:
            return fn
    return decorator
```

```python

slider = IntSlider(min=0, max=10, continuous_update=False)

x = np.linspace(0, 2 * np.pi, 1000)

out = Output(layout=Layout(height="100%"))

with out:
    fig = plt.figure(1)
    ax = fig.add_subplot(111, projection='polar')
    ax.plot(x, np.cos(x * 0))
    plt.show()


def plotx(change):
    with out:
        out.clear_output(wait=True)
        fig = plt.figure(1)
        ax = fig.add_subplot(111, projection='polar')
        ax.plot(x, np.cos(x * slider.value))
        plt.show()



# app = AppLayout(center=fig.canvas, header=slider, pane_heights=[1, 10, 0])

display(slider, out)

# plt.ion()
# plt.close()
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='polar')
# line, = ax.plot(x, np.cos(0 *x))

# out.display(ax.figure)
    

slider.observe(plotx, 'value')
```
